```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
data = pd.read_csv("all_data.csv")
data = data.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
data.head()
```

```python

means = data.drop("Year", axis = 1).groupby('Country').mean().reset_index()
means
```

```python
ax = plt.figure(figsize= (10,6))
sns.barplot(x= 'Country', y = 'LEABY', data = means, palette = 'magma')
plt.ylabel('Life expectancy at birth (years)')
```

```python
ax = plt.figure(figsize= (10,6))
sns.barplot(x= 'Country', y = 'GDP', data = means, palette = 'magma')
plt.ylabel('Life expectancy at birth (years)')
```
